Numt regions are specific genomic or mitochondrial regions where multiple numts are derived from. Practically it means the number of numt contributing nucleotides of the current region.

```python
#importing the required modules
import os
import pandas as pd
```

```python
master_array = pd.read_csv(r'Z:/balint/numt/Michael/data/numt_master_array.csv')
master_array.head()
```

```python
#genome parts
genome_parts = os.listdir(r'Z:\balint\numt\Michael\data\genome_parts_in_fasta/')
len(genome_parts)
```

```python
#genomic identifiers without repetition
g_ids = master_array['g_id'].unique()
g_ids
```

```python
#create sub dictionaries and dataframes for the different genomic ids
g_ids_dict = {}
for g_id in g_ids:
    g_ids_dict[g_id] = []
len(g_ids_dict)
```

```python
#get the genomic region corresponding values
for index, row in master_array.iterrows():
    g_id = row['g_id']
    if g_id in g_ids_dict:
        g_ids_dict[g_id].append(index)
```
