```python
#importing the modules
import os
import numpy as np
import pandas as pd
```

```python
home_dir=os.path.join(r'Z:\balint\numt/')
```

```python
numts=pd.read_csv(home_dir+r'Michael/data/numt_master_array.csv',
                 index_col=0)
numts.head()
```

```python
#get the cromosomes
chr_mask=numts['g_id'].apply(lambda g_id: len(g_id) < 3 and g_id!='MT')
chr_numts=numts[chr_mask]
chr_numts.head()
```

```python
mitos_annotation=pd.read_csv(home_dir+'mitos_annotation.bed',
                             sep='\t',
                             header=None,
                            index_col=0)
mitos_annotation.columns=['START','END','NAME','SIG','STRAND']
mitos_annotation.head()
```

```python
mt_ranges=mitos_annotation.apply(lambda row: set(np.arange(row['START'], row['END'], 1)), axis=1)
mt_ranges.index=mitos_annotation['NAME']
```

```python
mt_ranges
```

```python
numt_ranges=chr_numts.apply(lambda row: set(np.arange(row['mt_start'],
                                                             (row['mt_start']+row['mt_length']), 1)),
                                   axis=1)
numt_ranges.index=chr_numts['g_id']
(numt_ranges)
```

```python
#function for getting the interactions between chrs and mt parts
def common_nucleotides(numt_range):
    global interactions
    chr_id=numt_ranges[numt_ranges==numt_range].index[0]
    for mt_range in mt_ranges:
        mt_id=mt_ranges[mt_ranges==mt_range].index[0]
        if len(numt_range&mt_range) != 0:
            interactions.append(chr_id+','+mt_id)
```

```python
interactions=[]
numt_ranges.apply(common_nucleotides)
len(interactions)
```

```python
interactions_df=pd.DataFrame(0,
                             index=mt_ranges.index.values,
                            columns=np.arange(1,22,1))
interactions_df
```

```python
#function for filling the corresponding cells
def nuclotide_filler(interaction):
    chromosome_id=int(interaction.split(',')[0])
    mt_id=interaction.split(',')[1]
    interactions_df.loc[mt_id][chromosome_id]+=1
```

```python
pd.Series(interactions).apply(nuclotide_filler)
interactions_df
```

```python
interactions_df.to_csv(home_dir+'numts_chrs_interactions_nuclotides.csv')
```
