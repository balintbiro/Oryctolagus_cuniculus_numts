```python
import os
import pandas as pd
```

```python
home_dir = os.path.join(r'Z:/balint/numt/Michael/data/')
```

```python
numt_array = pd.read_csv(home_dir + r'numt_master_array_with_1000_flanking.csv')
numt_array = numt_array.drop(0)
numt_array
```

```python
len(numt_array['g_up_flanking'][2])
```

```python
#1000 flankings
upstream_flanking = numt_array['g_start'].apply(lambda start : start - 1000)
downstream_flanking = numt_array['g_start'].apply(lambda start : start + 1000)
flankings = pd.DataFrame([numt_array['g_id'], upstream_flanking, numt_array['g_start'],
                          downstream_flanking]).transpose()
flankings.columns = ['g_id','upstream','genome_start','downstream']
flankings.to_csv(home_dir + '1000_flankings.csv')
flankings
```

```python
for index, element in enumerate(upstream_flanking):
    print(str(index+1) + '_' + numt_array['g_id'][index+1] + ':' + str(element) + '-' + str(downstream_flanking[index + 1]))
```
