```python
numts=pd.read_csv(home_dir+'numt_master_array.csv')
numts.head()
```

```python
mt_annotation=pd.read_csv(home_dir+'mitos_annotation.bed',
                         sep='\t',
                         header=None,
                         index_col=0)
mt_annotation.columns=['start','end','name','sig','strand']
mt_annotation
```

```python
lengths=mt_annotation['end']-mt_annotation['start']
lengths.index=mt_annotation['name']
lengths
```

```python
#get the cromosomes
chr_mask=numts['g_id'].apply(lambda g_id: len(g_id) < 3 and g_id!='MT')
chr_numts=numts[chr_mask]
chr_numts=chr_numts.sort_values(by=['mt_length'], ascending=False)
chr_numts.head()
```

```python
def visualize_numts(row):
    global counter
    full_mt=17245+60
    size=0.015
    specified_color=color_dict[row['g_id']]
    numt_length=row['mt_length']
    leftover_mt=full_mt-numt_length
    startangle=row['mt_start']*(360/full_mt)
    ax.pie([leftover_mt, numt_length], radius=0.95-(counter*size), colors=['white',specified_color],
           wedgeprops=dict(width=size, edgecolor='w'), startangle=startangle)
    
    counter+=1
```

```python
counter=0
fig, ax = plt.subplots(figsize=(14, 14))
wedges, texts = ax.pie(lengths, wedgeprops=dict(width=0.05, edgecolor='w'), startangle=0,
                       labels=mt_annotation['name'], rotatelabels=True)
chr_numts.apply(visualize_numts, axis=1)
plt.savefig('../../results/mt_proba.png')
```
