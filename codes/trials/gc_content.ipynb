```python
#get the genomic ids
g_ids = numt_array['g_id']
g_ids = pd.Series(g_ids.unique())
```

```python
#get starts
starts = g_ids.apply(lambda x : numt_array.loc[numt_array['g_id'] == x]['g_start'].tolist())
starts.index = g_ids
starts
```

```python
#get ends
lengths = g_ids.apply(lambda x : numt_array.loc[numt_array['g_id'] == x]['g_length'].tolist())
lengths.index = g_ids
lengths
```

```python
#generate the ranges of numts eg.: that has to be excluded from the sampling
def range_to_exclude(g_id):
    starts_list = starts[g_id]
    lengths_list = lengths[g_id]
    to_exclude = []
    for start, length in zip(starts_list, lengths_list):
        exclude_range = list(np.arange(start, (start + (length + 1))))
        to_exclude += exclude_range
    return to_exclude
```

```python
ranges_to_exclude = g_ids.apply(range_to_exclude)
ranges_to_exclude.index = g_ids
ranges_to_exclude
```

```python
def get_sample_sequence(g_id):
    fasta_directory = os.path.join(r'D:\balint\numts\genome_parts_in_fasta/')
    filename = '%s.fasta' % g_id
    sequence = ''
    with open (fasta_directory + filename) as infile:
        content = infile.readlines()
        sequence = content[1]
    range_to_exclude = ranges_to_exclude[g_id]
    to_be_sampled = ''
    for index, nucleotide in enumerate(sequence):
        if index not in range_to_exclude:
            to_be_sampled += nucleotide
    global counter
    counter += 1
    print('done for %s chromosome(s)!' % counter)
    return to_be_sampled
```
