```python
#get the cromosomes
chr_mask=numts['g_id'].apply(lambda g_id: len(g_id) < 3 and g_id!='MT')
chr_numts=numts[chr_mask]
chr_numts.head()
```

```python
proba=chr_numts.loc[chr_numts['g_id']=='4']
proba
```

```python
proba=proba.sort_values(by=['g_start'])
proba
```

```python
connections=[]
starts=list(proba['g_start'])
genomic_size=list(proba['g_size'])[0]
g_lengths=list(proba['g_length'])
g_id=list(proba['g_id'])[0]
numt_id='proba'
for index, element in enumerate(starts):
    if index==0:
        self_link=element
        numt_length=g_lengths[index]
        connections.append([g_id, g_id, self_link, self_link])
        connections.append([numt_id, g_id, numt_length, numt_length])
    else:
        self_link=element-(starts[index-1]+g_lengths[index-1])
        numt_length=g_lengths[index]
        connections.append([g_id, g_id, self_link, self_link])
        connections.append([numt_id, g_id, numt_length, numt_length])
connections        
```
