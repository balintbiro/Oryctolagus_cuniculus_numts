```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('../data/e_values.txt')as e_vals,open('../data/identities.txt')as idents:
    e_values=e_vals.readlines()[0].split(',')[:-1]
    identities=idents.readlines()[0].split(',')[:-1]
```

```python
e_values=list(map(float,e_values))
identities=list(map(float,identities))
```

```python
plt.style.use('fivethirtyeight')
fig,axs=plt.subplots(1,2,sharex=True)
sns.boxplot(
    data=e_values,
    ax=axs[0],
    palette="Set3",
    width=0.7,
    linewidth=1,
    fliersize=0.0
)
axs[0].set_yscale('log')
sns.boxplot(
    data=identities,
    ax=axs[1],
    palette="Set3",
    width=0.7,
    linewidth=1,
    fliersize=0.0
)

sns.swarmplot(
        data=e_values,
        ax=axs[0],
        size=4,
        alpha=0.7
    )

sns.swarmplot(
        data=identities,
        ax=axs[1],
        size=4,
        alpha=0.7
    )
axs[0].set_ylabel('E value (log scale)')
axs[1].set_ylabel('Sequence identity (%)')
plt.tight_layout()
plt.savefig('../results/fpr_check.png',dpi=200)
```
