```python
#importing the required modules
import pandas as pd
import requests, sys
```

```python
#read the master_array into a dataframe
master_array = pd.read_csv(r'Z:/balint/numt/Michael/data/numt_master_array.csv')
master_array.head()
```

```python
#create function for obtaining data from Ensembl database
def get_ensembl_data(g_id, g_start, g_length):
    end = g_start + g_length
    ext = '/overlap/region/rabbit/%s:%s-%s?feature=gene' % (g_id, g_start, end)
    server = 'https://rest.ensembl.org'
    req = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
    if not req.ok:
        req.raise_for_status()
        sys.exit()
    output = req.json()
    #if len(output) != 0:
    #    gene_id = req.json()[0]['gene_id']
    #    return gene_id
    #else:
    #    return float('Nan')
    return output
```

```python
#get the gene ids for the specified genomic region
gene_ids = []
for index, row in master_array.iterrows():
    gene_id = get_ensembl_data(row['g_id'], row['g_start'], row['g_length'])
    #gene_ids.append(gene_id)
    print(gene_id)
```

```python
#update the master array
master_array['ensembl_gene_ids'] = gene_ids
```
